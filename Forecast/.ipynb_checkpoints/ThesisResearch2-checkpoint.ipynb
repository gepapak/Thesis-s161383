```python
# Import necessary libraries
import math
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import os
import ewtpy


from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras import layers
from tensorflow.keras.optimizers import Adam
from sklearn import metrics
from sklearn.svm import SVR
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.ensemble import RandomForestRegressor
from math import sqrt

path = "C:/Users/gepap/Desktop/Thesis/KarijadiLocal/scenario_000.csv"
df = pd.read_csv(path)

#convert an array of values into a dataset matrix
def create_dataset(dataset, look_back=1):
    dataX, dataY = [], []
    for i in range(len(dataset)-look_back-1):
        a = dataset[i:(i+look_back), 0]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])
    return np.array(dataX), np.array(dataY)

df

df['Date'] = pd.to_datetime(df['timestamp'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
new_data=df[['Month','Year','Date','wind']]
# new_data=new_data[new_data.Year == 2020]

cap=max(new_data['wind'])

i=[1,2] #for fold-1
look_back=6
data_partition=0.5257331291956189

CUDA_VISIBLE_DEVICES=""
from PyEMD import CEEMDAN

# ANN Model Function
def ann_model(new_data, i, look_back, data_partition, cap):
    data1 = new_data.loc[new_data['Month'].isin(i)].reset_index(drop=True).dropna()
    datas = data1['wind']
    datasets = datas.values.reshape(-1, 1)

    train_size = int(len(datasets) * data_partition)
    train, test = datasets[:train_size], datasets[train_size:]

    trainX, trainY = create_dataset(train, look_back)
    testX, testY = create_dataset(test, look_back)

    sc_X = StandardScaler()
    sc_y = StandardScaler()

    X_train = sc_X.fit_transform(trainX)
    y_train = sc_y.fit_transform(trainY.reshape(-1, 1)).ravel()
    X_test = sc_X.transform(testX)
    y_test = sc_y.transform(testY.reshape(-1, 1)).ravel()

    # Ensure data is float32 for TensorFlow
    X_train = np.array(X_train, dtype=np.float32)
    y_train = np.array(y_train, dtype=np.float32)
    X_test = np.array(X_test, dtype=np.float32)
    y_test = np.array(y_test, dtype=np.float32)

    # ✅ Keep 2D shape (no extra dimension)
    trainX1 = X_train
    testX1 = X_test

    # Set random seed for reproducibility
    np.random.seed(1234)
    tf.random.set_seed(1234)
    os.environ['PYTHONHASHSEED'] = '0'
    tf.config.experimental.enable_op_determinism()

    # ✅ Correct Input Shape for ANN
    neuron = 128
    model = Sequential()
    model.add(Dense(units=neuron, activation='relu', input_shape=(trainX1.shape[1],)))  # Fixed input shape
    model.add(Dense(1))

    model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam())
    model.fit(trainX1, y_train, verbose=0)

    # Make predictions
    y_pred_train = model.predict(trainX1)
    y_pred_test = model.predict(testX1).ravel()

    # Inverse transform predictions
    y_pred_test = sc_y.inverse_transform(y_pred_test.reshape(-1, 1))
    y_test = sc_y.inverse_transform(y_test.reshape(-1, 1))

    # Summarize the fit of the model
    mape = np.mean((np.abs(y_test - y_pred_test)) / cap) * 100
    rmse = sqrt(mean_squared_error(y_test, y_pred_test))
    mae = mean_absolute_error(y_test, y_pred_test)

    print('MAPE:', mape)
    print('RMSE:', rmse)
    print('MAE:', mae)
    
def proposed_method_global_scaling(new_data, i, look_back, data_partition, cap):
    from tensorflow.keras.models import Sequential
    from tensorflow.keras.layers import LSTM, Dense
    from sklearn.preprocessing import StandardScaler
    from sklearn.metrics import mean_squared_error, mean_absolute_error
    from math import sqrt
    import numpy as np
    import pandas as pd
    import os
    import tensorflow as tf
    import ewtpy
    from PyEMD import CEEMDAN
    import gc
    from tensorflow.keras import backend as K

    # Step 1: Extract the relevant data and apply global scaling
    data1 = new_data.loc[new_data['Month'].isin(i)].reset_index(drop=True).dropna()
    original_series = data1['wind'].values.reshape(-1, 1)

    # Global scaling on full original signal
    scaler = StandardScaler()
    scaled_series = scaler.fit_transform(original_series)

    # Step 2: CEEMDAN decomposition
    emd = CEEMDAN(epsilon=0.05)
    emd.noise_seed(12345)
    IMFs = emd(scaled_series.flatten())
    ceemdan1 = pd.DataFrame(IMFs).T
    imf1 = ceemdan1.iloc[:, 0]

    # Step 3: EWT on IMF1
    ewt, _, _ = ewtpy.EWT1D(imf1, N=3)
    df_ewt = pd.DataFrame(ewt)
    if df_ewt.shape[1] > 2:
        df_ewt.drop(columns=[2], inplace=True)
    denoised = df_ewt.sum(axis=1)

    # Step 4: Rebuild component matrix (denoised + other IMFs)
    ceemdan_without_imf1 = ceemdan1.iloc[:, 1:]
    new_ceemdan = pd.concat([denoised, ceemdan_without_imf1], axis=1)

    print("new_ceemdan shape:", new_ceemdan.shape)

    pred_test, pred_train = [], []

    # Step 5: Train LSTM for each IMF component
    for col in new_ceemdan:
        dataset = new_ceemdan[[col]].values

        train_size = int(len(dataset) * data_partition)
        train, test = dataset[:train_size], dataset[train_size:]

        def create_dataset(dataset, look_back):
            dataX, dataY = [], []
            for i in range(len(dataset) - look_back - 1):
                a = dataset[i:(i + look_back), 0]
                dataX.append(a)
                dataY.append(dataset[i + look_back, 0])
            return np.array(dataX), np.array(dataY)

        trainX, trainY = create_dataset(train, look_back)
        testX, testY = create_dataset(test, look_back)

        trainX = trainX.reshape(trainX.shape[0], look_back, 1)
        testX = testX.reshape(testX.shape[0], look_back, 1)

        # Model
        model = Sequential()
        model.add(LSTM(128, input_shape=(look_back, 1)))
        model.add(Dense(1))
        model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(learning_rate=0.001))
        model.fit(trainX, trainY, epochs=100, batch_size=64, verbose=0)

        y_pred_train = model.predict(trainX).ravel()
        y_pred_test = model.predict(testX).ravel()

        pred_train.append(y_pred_train)
        pred_test.append(y_pred_test)

        # Memory cleanup
        K.clear_session()
        del model
        gc.collect()

    # Step 6: Sum across IMF predictions
    pred_test = np.array(pred_test).reshape(len(pred_test), -1)
    pred_train = np.array(pred_train).reshape(len(pred_train), -1)
    summed_pred_test = pred_test.sum(axis=0)
    summed_pred_train = pred_train.sum(axis=0)

    # Step 7: Inverse transform the final summed predictions
    summed_pred_test_inv = scaler.inverse_transform(summed_pred_test.reshape(-1, 1))
    summed_pred_train_inv = scaler.inverse_transform(summed_pred_train.reshape(-1, 1))

    # Reconstruct actual y_test from original scaled series
    _, actual_y_test = create_dataset(scaled_series[int(len(scaled_series) * data_partition):], look_back)
    actual_y_test_inv = scaler.inverse_transform(actual_y_test.reshape(-1, 1))

    # Metrics
    mape = np.mean(np.abs(actual_y_test_inv - summed_pred_test_inv) / cap) * 100
    rmse = sqrt(mean_squared_error(actual_y_test_inv, summed_pred_test_inv))
    mae = mean_absolute_error(actual_y_test_inv, summed_pred_test_inv)

    print("MAPE:", mape)
    print("RMSE:", rmse)
    print("MAE:", mae)

    return pd.DataFrame(summed_pred_test_inv), pd.DataFrame(summed_pred_train_inv), pd.DataFrame(actual_y_test_inv)
```

```python
import numpy as np
import pandas as pd
from PyEMD import CEEMDAN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import tensorflow as tf
import gc

LOOK_BACK = 6
DATA_PARTITION = 0.7

# Load your file
df = pd.read_csv("scenario_000.csv")
df['Date'] = pd.to_datetime(df['timestamp'])

# Select features
features = df[["wind", "market_price", "load"]].values
target = df[["wind"]].values
cap = df["wind"].max()

# Scaling all variables
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# Apply CEEMDAN to each variable
emd = CEEMDAN(epsilon=0.05)
emd.noise_seed(123)

IMF_list = []
for i in range(features_scaled.shape[1]):
    IMFs = emd(features_scaled[:, i])
    IMF_list.append(IMFs)

# Ensure same number of IMFs
min_imfs = min([imfs.shape[0] for imfs in IMF_list])
IMF_list = [imfs[:min_imfs] for imfs in IMF_list]

# Stack as multi-variate input
X_all = np.hstack([imfs.T for imfs in IMF_list])

# Prepare data
def create_dataset(X, y, look_back):
    Xs, ys = [], []
    for i in range(len(X) - look_back):
        Xs.append(X[i:i+look_back])
        ys.append(y[i + look_back])
    return np.array(Xs), np.array(ys)

train_size = int(len(X_all) * DATA_PARTITION)
X_train, y_train = create_dataset(X_all[:train_size], target[:train_size], LOOK_BACK)
X_test, y_test = create_dataset(X_all[train_size:], target[train_size:], LOOK_BACK)

# Model
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape=(LOOK_BACK, X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(64))
model.add(Dense(1))
model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(learning_rate=0.001))

model.fit(X_train, y_train, epochs=100, batch_size=64, verbose=1)

y_pred = model.predict(X_test)

# Inverse Scale Wind Prediction
wind_idx = 0  # First column is wind
X_dummy = np.zeros((len(y_pred), features.shape[1]))
X_dummy[:, wind_idx] = y_pred.flatten()
y_pred_inv = scaler.inverse_transform(X_dummy)[:, wind_idx]

y_test_inv = y_test.flatten()

NRMSE = np.sqrt(mean_squared_error(y_test_inv, y_pred_inv)) / cap
NMAE = mean_absolute_error(y_test_inv, y_pred_inv) / cap

print("NRMSE:", NRMSE)
print("NMAE:", NMAE)

gc.collect()
```

```python
# === MV-EEMD + LSTM for Wind, Solar, Hydro Forecasting ===

import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import os
import gc

from PyEMD import EEMD
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
from tensorflow.keras import backend as K
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt


path = "C:/Users/gepap/Desktop/Thesis/KarijadiLocal/scenario_000.csv"
df = pd.read_csv(path)

df['Date'] = pd.to_datetime(df['timestamp'])
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

new_data = df[['Month', 'Year', 'Date', 'wind', 'solar', 'hydro']]

cap = max(new_data['wind'])
look_back = 6
data_partition = 0.52

def create_dataset_mv(dataset, look_back):
    dataX, dataY = [], []
    for i in range(len(dataset) - look_back - 1):
        a = dataset[i:(i + look_back), :]
        dataX.append(a)
        dataY.append(dataset[i + look_back, 0])  # Forecasting wind only
    return np.array(dataX), np.array(dataY)


def mv_eemd_lstm(new_data, look_back, data_partition, cap):
    data = new_data.drop(columns=['Month', 'Year', 'Date']).values
    scaler = StandardScaler()
    scaled_data = scaler.fit_transform(data)

    n_components = []

    # EEMD on each variable separately
    for i in range(scaled_data.shape[1]):
        signal = scaled_data[:, i]
        eemd = EEMD()
        imfs = eemd.eemd(signal)
        print(f"Variable {i+1} decomposed into {imfs.shape[0]} IMFs")
        n_components.append(imfs)

    # Align IMFs across all variables
    min_imfs = min(imf.shape[0] for imf in n_components)
    print(f"Using {min_imfs} IMFs for all variables")

    results_test, results_train = [], []

    for imf_idx in range(min_imfs):
        imf_signals = []
        for var_imf in n_components:
            imf_signals.append(var_imf[imf_idx, :])

        component = np.array(imf_signals).T  # shape (samples, variables)

        train_size = int(len(component) * data_partition)
        train, test = component[:train_size], component[train_size:]

        trainX, trainY = create_dataset_mv(train, look_back)
        testX, testY = create_dataset_mv(test, look_back)

        model = Sequential()
        model.add(LSTM(128, input_shape=(look_back, trainX.shape[2]), return_sequences=False))
        model.add(Dropout(0.2))
        model.add(Dense(1))
        model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='mse')

        model.fit(trainX, trainY, epochs=100, batch_size=64, verbose=0)

        y_pred_train = model.predict(trainX).ravel()
        y_pred_test = model.predict(testX).ravel()

        results_train.append(y_pred_train)
        results_test.append(y_pred_test)

        K.clear_session()
        del model
        gc.collect()

    summed_train = np.sum(results_train, axis=0)
    summed_test = np.sum(results_test, axis=0)

    _, actual_y_test = create_dataset_mv(scaled_data[int(len(scaled_data) * data_partition):], look_back)

    summed_test_inv = scaler.inverse_transform(np.column_stack([summed_test] + [np.zeros_like(summed_test)] * (scaled_data.shape[1] - 1)))[:, 0]
    actual_y_test_inv = scaler.inverse_transform(np.column_stack([actual_y_test] + [np.zeros_like(actual_y_test)] * (scaled_data.shape[1] - 1)))[:, 0]

    mape = np.mean(np.abs(actual_y_test_inv - summed_test_inv) / cap) * 100
    rmse = sqrt(mean_squared_error(actual_y_test_inv, summed_test_inv))
    mae = mean_absolute_error(actual_y_test_inv, summed_test_inv)

    print(f"MAPE: {mape}")
    print(f"RMSE: {rmse}")
    print(f"MAE: {mae}")

    return summed_test_inv, summed_train, actual_y_test_inv


summed_test, summed_train, actual_y_test = mv_eemd_lstm(new_data, look_back, data_partition, cap)
```

```python
# === Wavelet + CEEMDAN + LSTM Hybrid for Wind Forecasting ===

import numpy as np
import pandas as pd
import pywt
from PyEMD import CEEMDAN
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
from math import sqrt
import gc
import matplotlib.pyplot as plt

# === CONFIG ===
LOOK_BACK = 6
DATA_PARTITION = 0.7
WAVELET = 'db4'

# Load your scenario
df = pd.read_csv("scenario_000.csv")

df['Date'] = pd.to_datetime(df['timestamp'])
features = df[["wind", "market_price", "load"]].values
target = df[["wind"]].values
cap = df["wind"].max()

# === Scaling ===
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

# === CEEMDAN Decomposition per variable ===
emd = CEEMDAN(epsilon=0.05)
emd.noise_seed(123)

IMF_list = []
for i in range(features_scaled.shape[1]):
    imfs = emd(features_scaled[:, i])
    IMF_list.append(imfs)

min_imfs = min([imfs.shape[0] for imfs in IMF_list])
IMF_list = [imfs[:min_imfs] for imfs in IMF_list]

# === Apply Wavelet Denoising on IMF1 of each variable ===
denoised_imf1 = []
for imfs in IMF_list:
    coeffs = pywt.wavedec(imfs[0], WAVELET, level=3)
    threshold = np.median(np.abs(coeffs[-1])) / 0.6745 * np.sqrt(2 * np.log(len(imfs[0])))
    coeffs_denoised = [pywt.threshold(c, threshold, mode='soft') for c in coeffs]
    clean_signal = pywt.waverec(coeffs_denoised, WAVELET)
    clean_signal = clean_signal[:len(imfs[0])]
    denoised_imf1.append(clean_signal)

# === Rebuild Components ===
X_all = np.hstack([clean_signal.reshape(-1, 1) for clean_signal in denoised_imf1])

# === Prepare Dataset ===
def create_dataset(X, y, look_back):
    Xs, ys = [], []
    for i in range(len(X) - look_back):
        Xs.append(X[i:i + look_back])
        ys.append(y[i + look_back])
    return np.array(Xs), np.array(ys)

train_size = int(len(X_all) * DATA_PARTITION)
X_train, y_train = create_dataset(X_all[:train_size], target[:train_size], LOOK_BACK)
X_test, y_test = create_dataset(X_all[train_size:], target[train_size:], LOOK_BACK)

# === LSTM Model ===
model = Sequential()
model.add(LSTM(128, return_sequences=True, input_shape=(LOOK_BACK, X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(64))
model.add(Dense(1))
model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(learning_rate=0.001))

model.fit(X_train, y_train, epochs=100, batch_size=64, verbose=1)

# === Prediction ===
y_pred = model.predict(X_test)

# === Inverse Transform Wind ===
wind_idx = 0
dummy = np.zeros((len(y_pred), features.shape[1]))
dummy[:, wind_idx] = y_pred.flatten()

y_pred_inv = scaler.inverse_transform(dummy)[:, wind_idx]
y_test_inv = y_test.flatten()

# === Metrics ===
RMSE = sqrt(mean_squared_error(y_test_inv, y_pred_inv))
MAE = mean_absolute_error(y_test_inv, y_pred_inv)
NRMSE = RMSE / cap
NMAE = MAE / cap

print("RMSE:", RMSE)
print("MAE:", MAE)
print("NRMSE:", NRMSE)
print("NMAE:", NMAE)

gc.collect()

# === Visualization ===
plt.figure(figsize=(12, 4))
plt.plot(y_test_inv[:200], label='Actual')
plt.plot(y_pred_inv[:200], label='Predicted')
plt.legend()
plt.title("Wavelet + CEEMDAN + LSTM Forecast (Wind)")
plt.show()
```

```python
import numpy as np
import pandas as pd
from PyEMD import CEEMDAN
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error
import tensorflow as tf
import gc
from math import sqrt

# Config
LOOK_BACK = 6
DATA_PARTITION = 0.7

# Load Data
df = pd.read_csv("scenario_000.csv")
df['Date'] = pd.to_datetime(df['timestamp'])
features = df[['wind', 'market_price', 'load']].values
target = df['wind'].values.reshape(-1, 1)
cap = df['wind'].max()

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)

emd = CEEMDAN(epsilon=0.05)
emd.noise_seed(123)

# Decompose each variable separately
IMF_list = []
for i in range(features_scaled.shape[1]):
    imfs = emd(features_scaled[:, i])
    IMF_list.append(imfs)

min_imfs = min([imfs.shape[0] for imfs in IMF_list])
print(f"Using {min_imfs} IMFs")

# Align & Concatenate
aligned_imfs = [imfs[:min_imfs, :].T for imfs in IMF_list]
X_all = np.hstack(aligned_imfs)

# Dataset Creator
def create_dataset(X, y, look_back):
    Xs, ys = [], []
    for i in range(len(X) - look_back):
        Xs.append(X[i:i+look_back])
        ys.append(y[i + look_back])
    return np.array(Xs), np.array(ys)

train_size = int(len(X_all) * DATA_PARTITION)
X_train, y_train = create_dataset(X_all[:train_size], target[:train_size], LOOK_BACK)
X_test, y_test = create_dataset(X_all[train_size:], target[train_size:], LOOK_BACK)

# Model
model = Sequential([
    LSTM(128, input_shape=(LOOK_BACK, X_train.shape[2]), return_sequences=True),
    Dropout(0.2),
    LSTM(64),
    Dense(1)
])
model.compile(loss='mse', optimizer=tf.keras.optimizers.Adam(learning_rate=0.001))
model.fit(X_train, y_train, epochs=100, batch_size=64, verbose=1)

# Predict
y_pred = model.predict(X_test)

# Inverse Scaling (only wind)
X_dummy = np.zeros((len(y_pred), features.shape[1]))
X_dummy[:, 0] = y_pred.flatten()
y_pred_inv = scaler.inverse_transform(X_dummy)[:, 0]

# True Values
y_test_inv = y_test.flatten()

print("RMSE:", sqrt(mean_squared_error(y_test_inv, y_pred_inv)))
print("MAE:", mean_absolute_error(y_test_inv, y_pred_inv))
print("NRMSE:", sqrt(mean_squared_error(y_test_inv, y_pred_inv)) / cap)
print("NMAE:", mean_absolute_error(y_test_inv, y_pred_inv) / cap)

gc.collect()
```
